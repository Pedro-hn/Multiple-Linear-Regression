##Load necessary libraries

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.metrics import r2_score
```

```python
df1 = pd.read_csv("train (2).csv")
df2 = pd.read_csv("test (2).csv")
```

```python
df1.head()
```

```python
df2.head()
```

```python
df = pd.concat([df1, df2], ignore_index=True, axis=0)
```

```python
df.duplicated().sum()
```

```python
analysis_one = pd.DataFrame(df.isna().sum())
analysis_one
```

Treating NA missing values

```python
mean_one = df["LotFrontage"].mean()
```

```python
df['LotFrontage'] = df['LotFrontage'].fillna(mean_one)
```

But we won't use all the variables in this case!!

```python
df_y = df["SalePrice"]
```

```python
df = df.iloc[:,0:6]
df.head()
```

```python
df = df.dropna()
```

```python
df.shape
```

```python
df_y.head()
```

```python
df.iloc[:,1:6]
```

##One Hot Encoder

```python
df.dtypes
```

```python
df["MSZoning"].unique()
```

```python
df["Street"].unique()
```

```python
ohe = OneHotEncoder()
```

```python
feature_arry = ohe.fit_transform(df[["Street"]]).toarray()
feature_arry
```

```python
ohe.categories_
```

```python
feature_labels = ohe.categories_
feature_labels
```

```python
feature_labels = np.array(feature_labels).ravel()
print(feature_labels)
```

```python
street_ohe = pd.DataFrame(feature_arry, columns = feature_labels)
```

```python
df.head()
```

```python
df = df.drop(["MSZoning"], axis = 1)
```

```python
df_final = pd.concat([df, street_ohe], axis = 1)
```

```python
df_final.head()
```

```python
df = df_final.drop(["Street"], axis =1)
df.head()
```

##Splitting the dataset in train and test

---



```python
X = df.iloc[:,1:6]
y = df_y
```

```python
X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X.iloc[1:80,], y.iloc[1:80,], test_size=.3)
```

##Multiple Linear Regression

```python
reg = linear_model.LinearRegression()
reg.fit(X_train, y_train)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
y_predict = reg.predict(X_test)
```

##Evaluating the model

```python
r2_score(y_test, y_predict)
```
